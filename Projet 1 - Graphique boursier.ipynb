```python
# 1er graphique : Visuel de l'action Apple via un graphique en bougies japonaises, un graphique linéraire ainsi que d'un démontrant les volumes.
# 1st chart: Visualization of the Apple share via a Japanese candlestick chart, a line chart and a volume chart.

import pandas as pd 
import mplfinance as mpf 

data = pd.read_csv("/Users/massinissaalouache/Downloads/AAPL.csv")
data["Date"] = pd.to_datetime(data["Date"])
data.set_index("Date", inplace=True)

mpf.plot(data, type="candle", style="yahoo")
mpf.plot(data, type="line", volume=True)
```

```python
# 2nd graphique : Visuel de l'action Apple via un graphique en bougies japonaises, un graphique linéraire ainsi que d'un démontrant les volumes sur le mois d'Avril 2023.
# 2nd chart: Visual of Apple shares via a Japanese candlestick chart, a line chart and a volume chart for April 2023.

import pandas as pd
import mplfinance as mpf

data = pd.read_csv("/Users/massinissaalouache/Downloads/AAPL.csv")
data["Date"] = pd.to_datetime(data["Date"])
data.set_index("Date", inplace=True)

mpf.plot(data.loc["2023-04"], type="candle", style="yahoo")
mpf.plot(data.loc["2023-04"], type="line", volume=True)# 2) 
```

```python
# 3ème graphique : Visuel de l'action Apple via un graphique en bougies japonaises du 1er janvier 2023 au 1er septembre 2023. Ce graphique est complété par un indicateur de 
# Volumes et un indicateur de Moyenne Mobile.

# 3rd chart: Visualization of the Apple share via a Japanese candlestick chart from January 1, 2023 to September 1, 2023. This chart is completed by a Volumes and a 
# Moving Average indicator.

import pandas as pd
import mplfinance as mpf

data = pd.read_csv("/Users/massinissaalouache/Downloads/AAPL.csv")
data["Date"] = pd.to_datetime(data["Date"])
data.set_index("Date", inplace=True)

mpf.plot(data.loc["2023-01" : "2023-07"], type="candle", volume= True, mav = (7), style="yahoo")
```
