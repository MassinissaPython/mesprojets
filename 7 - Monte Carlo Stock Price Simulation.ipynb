```python
# EN : Stock price simulation using Monte Carlo methods 
# FR : Simulation de prix d'une action avec la methode de Monte Carlo 
```

```python
import yfinance as yf 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
df = yf.download("AAPL")
returns = np.log(1+df["Adj Close"].pct_change())
mu, sigma = returns.mean(), returns.std()
simulation_returns = np.random.normal(mu,sigma, 252)
intiial_price = df["Adj Close"].iloc[-1]
simulation_price = intiial_price * (simulation_returns + 1).cumprod()
```

```python
plt.plot(simulation_price)
```

```python
for i in range(100):
    simulation_returns = np.random.normal(mu,sigma, 252)
    simulation_price = intiial_price * (simulation_returns + 1).cumprod() 
    plt.axhline(intiial_price, c="k")
    plt.plot(simulation_price)
```
