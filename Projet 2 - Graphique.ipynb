```python
import pandas as pd
import mplfinance as mpf
import matplotlib.pyplot as plt
```

```python
# Graphique boursier Coca Cola (KO)
# Coca Cola stock chart (KO)

data_ko = pd.read_csv("/Users/massinissaalouache/Downloads/KO.csv")
data_ko["Date"] = pd.to_datetime(data_ko["Date"])
data_ko.set_index("Date", inplace=True)

mpf.plot(data_ko.loc["2023-02": "2023-07"], type="candle", style="yahoo", mav=(8), volume=True)
```

```python
# Graphique en nuage de points entre le volume et le prix 
# Scatter plot between volume and price 

plt.scatter(data_ko["Close"], data_ko["Volume"], s=5, c=data_ko["Close"], cmap="inferno")
plt.xlabel("Close")
plt.ylabel("Volume")
plt.title("Close/Volume scatter chart")
plt.show()
```

```python
# Histogramme des volumes 

plt.hist(data_ko.loc[data_ko["Volume"]> 10]["Volume"], bins=100)
plt.show()
```

```python
# Graphique boursier Bitcoin-Euro (BTC-EUR) 
# BTC-EUR stock chart 

data_btc_eur = pd.read_csv("/Users/massinissaalouache/Downloads/BTC-EUR.csv")
data_btc_eur["Date"] = pd.to_datetime(data_btc_eur["Date"])
data_btc_eur.set_index("Date", inplace=True)

mpf.plot(data_btc_eur.loc["2023-10"], type="candle", style="yahoo",mav=(7), volume=True)
```

```python
# Graphique en nuage de points entre le volume et le close de l'actif
# Scatter plot between asset volume and asset close

plt.scatter(data_btc_eur["Volume"], data_btc_eur["Close"], s=5, c=data_btc_eur["Volume"], cmap="inferno")
plt.xlabel("Volume")
plt.ylabel("Close")
plt.title("Volume/Close scatter chart")
plt.show
```
