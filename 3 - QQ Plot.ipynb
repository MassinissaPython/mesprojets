```python
# En : 
# 2 Ways to test for normality of returns (with python code!)
# Q-Q plot - A Quantile-Quantile (QQ) plot is a visual method for comparing a sample to a theoretical distribution. On the x-axis, we plot the theoretical quantiles (in this case, those of a standard normal distribution), and on the y-axis, we plot the sample quantiles. If the distributions are similar, the resulting QQ plot will be linear. In other words, if we observe that the QQ plot forms a straight line, we can safely conclude that the returns follow a normal distribution.
# Jarque-Bera Test - This is a statistical test used to check whether a given sample is normally distributed or not. According to the scipy.stats module documentation, the Jarque-Bera test assesses whether the sample data has skewness and kurtosis matching a normal distribution. If the p-value is less than the chosen significance level, it indicates that the returns are not normally distributed. If the p-value is greater than the chosen significance level, it suggests that the returns are normally distributed.
# In this context, we calculated discrete proportional returns for Tesla stock over a 2-year lookback period. Based on a Q-Q plot and the Jarque-Bera test, we can conclude that the returns are not normally distributed.
# What are your thoughts about the same? Let me know in the comments. By Shreyas Ingale on Linkedin !

# Fr : 
# Le QQ Plot (Quantile-Quantile Plot) en finance est un graphique de dispersion utilisé pour évaluer la distribution de données, en particulier pour vérifier si une série temporelle de rendements suit une distribution normale. Il compare les quantiles théoriques d'une distribution avec les quantiles observés des données.
# En finance, le QQ Plot returns est souvent utilisé pour visualiser la normalité des rendements financiers. Si les points du graphique s'alignent approximativement sur une ligne droite, cela suggère que les rendements suivent une distribution normale. Cependant, des déviations de cette ligne droite peuvent indiquer des non-normalités dans la distribution des rendements.
# Il est essentiel de noter que le QQ Plot est un outil visuel, et son interprétation nécessite une compréhension approfondie des concepts statistiques. En finance, il est couramment utilisé pour évaluer l'adéquation des rendements aux hypothèses de modèles financiers.
```

```python
import pandas as pd
import yfinance as yf
from datetime import date, timedelta
import statsmodels.api as sm
import matplotlib.pyplot as plt
from scipy.stats import jarque_bera
```

```python
ticker = "TSLA"
end = date.today()
start = date.today() - timedelta(10*365)

tsla = yf.download(ticker, start, end) ["Close"]
tsla_rets = tsla.pct_change().dropna()

sm.qqplot(tsla_rets.values, line = "s")
plt.title("Q-Q plot for discret returns of TSLA returns")
plt.grid
plt.show
```
